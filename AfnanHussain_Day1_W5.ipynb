```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
df = pd.DataFrame(columns=['CustomerID','Age','Gender','MartialStatus','AnnualIncome (USD)','TotalPurchases','PreferredCategory'])
df
```

```python
id = [1001,1002,1003,1004,1005]
age = [33,28,42,51,37]
gender = ['Male','Female','Male','Female','Male']
status = ['Married','Single','Married','Single','Divorced']
income = [65000,45000,55000,80000,58000]
purchases = [18,15,20,12,10]
category = ['Electronics','Appliances','Electronics','Electronics','Appliances']
```

```python
random.seed(42)
app_choice = ['Electronics','Appliances']
status_choice = ['Married','Single','Divorced']
gender_choice = ['Male','Female']
for i in range(6,501):
  num = random.randint(0,2)
  Age = random.randint(15,80)
  g = random.choice(gender_choice)
  s = random.choice(status_choice)
  Income = random.randint(10000,100000)
  purchase = random.randint(1,100)
  cat = random.choice(app_choice)
  id.append(1000+i)
  age.append(Age)
  gender.append(g)
  income.append(Income)
  purchases.append(purchase)
  category.append(cat)
  status.append(s)
```

```python
df['CustomerID'] = id
df['Age'] = age
df['Gender'] = gender
df['MartialStatus'] = status
df['AnnualIncome (USD)'] = income
df['TotalPurchases'] = purchases
df['PreferredCategory'] = category
df
```

```python
df.isna().sum()
```

```python
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
for col in df.columns:
  if df[col].dtype == 'int' or df[col].dtype == 'float64':
    outliers = find_outliers_IQR(df[col])
    print("Number of outliers in",col,":",len(outliers))
```

```python
from sklearn.cluster import KMeans
```

```python
inertias = []
features_for_segmentation = ['Age', 'AnnualIncome (USD)', 'TotalPurchases']
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  model = kmeans.fit(df[features_for_segmentation])
  inertias.append(model.inertia_)

plt.plot(range(1, 11), inertias, marker='.')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Cluster Selection')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[features_for_segmentation])
sns.scatterplot(data=df, x='Age', y='TotalPurchases', hue='Cluster', palette='Set1')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 2], s=200, c='red', marker='X', label='Cluster Centers')
plt.title('Customer Segmentation')
plt.xlabel('Age')
plt.ylabel('Total Purchases')
plt.legend()
plt.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from plotly.tools import mpl_to_plotly
```

```python
# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1('TechElectro Customer Segmentation Dashboard'),

    # Dropdown for selecting the number of clusters
    html.Label('Select the number of clusters:'),
    dcc.Dropdown(
        id='cluster-dropdown',
        options=[
            {'label': str(i), 'value': i} for i in range(1, 11)
        ],
        value=3,
        clearable=False
    ),

    # Scatter plot for customer segmentation
    dcc.Graph(id='scatter-plot'),

    # Histogram for Annual Income distribution
    dcc.Graph(id='histogram-annual-income')
])

# Define callback functions to update the dashboard components
@app.callback(
    Output('scatter-plot', 'figure'),
    Output('histogram-annual-income', 'figure'),
    Input('cluster-dropdown', 'value')
)
def update_graphs(selected_clusters):
    # Apply K-means clustering with the selected number of clusters
    kmeans = KMeans(n_clusters=selected_clusters, random_state=42)
    df['Cluster'] = kmeans.fit_predict(df[features_for_segmentation])

    # Scatter plot for customer segmentation
    scatter_fig = sns.scatterplot(data=df, x='Age', y='TotalPurchases', hue='Cluster', palette='Set1').get_figure()
    scatter_fig.suptitle('Customer Segmentation', fontsize=16)
    plt.xlabel('Age')
    plt.ylabel('Total Purchases')
    scatter_plotly = mpl_to_plotly(scatter_fig)  # Convert matplotlib figure to Plotly figure

    # Histogram for Annual Income distribution
    hist_fig = sns.histplot(df['AnnualIncome (USD)'], bins=20, kde=True).get_figure()
    hist_fig.suptitle('Distribution of Annual Income', fontsize=16)
    plt.xlabel('Annual Income (USD)')
    plt.ylabel('Count')
    histogram_plotly = mpl_to_plotly(hist_fig)  # Convert matplotlib figure to Plotly figure

    return scatter_plotly, histogram_plotly

if __name__ == '__main__':
    app.run_server(debug=True)
```
