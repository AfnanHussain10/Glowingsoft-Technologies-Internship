# **Day 3 Task 1**
### Website Scraping

```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import Request, urlopen
import seaborn as sns
```

```python
URL = "https://www.daftlogic.com/information-appliance-power-consumption.htm"
print(URL)
```

```python
req = Request( url=URL, headers={'User-Agent': 'Chrome/114.0'})
page = urlopen(req).read()
page
title = str(page).split('<title>')[1].split('</title>')[0]
title
```

```python
soup = BeautifulSoup(page, 'html.parser')
table = soup.find('table', id="tblApp")
table
```

```python
df = pd.read_html(str(table))[0]
df.head()
```

```python
df.to_csv("appliancesInfo.csv")
data=pd.read_csv("appliancesInfo.csv")
data.head()
```
